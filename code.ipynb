```python
import pandas as pd
```

```python
data = pd.read_csv('data.csv', sep=';', engine='python', encoding='utf-8')
```

```python
print("Informações gerais sobre o conjunto de dados:")
display(data.info())
```

```python
print("Primeiras 5 linhas:")
display(data.head())
```

```python
print("Últimas 5 linhas:")
display(data.tail())
```

```python
data_copia = data.copy()
```

```python
data_copia['Calories'] = data_copia['Calories'].fillna(0)
print("Dados após substituir nulos em 'Calories':")
display(data_copia)
```

```python
data_copia['Date'] = data_copia['Date'].fillna('1900/01/01')
print("Dados após substituir nulos em 'Date':")
display(data_copia)
```

```python
data_copia['Date'] = pd.to_datetime(data_copia['Date'], errors='coerce')
print("Dados após a conversão inicial para datetime:")
display(data_copia)
```

```python
data_copia['Date'] = data_copia['Date'].replace('1900-01-01', pd.NaT)
print("Dados após substituir '1900/01/01' por NaN:")
display(data_copia)
```

```python
data_copia['Date'] = pd.to_datetime(data_copia['Date'], errors='coerce')
print("Dados após a nova conversão para datetime:")
display(data_copia)
```

```python
data_copia['Date'] = data_copia['Date'].replace('20201226', '2020/12/26')
```

```python
data_copia['Date'] = pd.to_datetime(data_copia['Date'], errors='coerce')
print("Dados após corrigir '20201226' e converter para datetime:")
display(data_copia)
```

```python
data_copia.dropna(subset=['Date'], inplace=True)
```

```python
print("Dados finais após todas as transformações e remoção de nulos:")
display(data_copia)
```
